```python
#Extracting individual columns
print(df['total_bill'])
print(df.total_bill)        #Another syntax
print(df[['total_bill','tip']])
```

```python
#Add a new column
df['tip_percent'] = 100 * df['tip'] / df['total_bill']
df['price_per_person'] = df['total_bill'] /df['size']
print(df.head())
df['price_per_person'] = np.round(df['price_per_person'],2)
print(df.head())
```

```python
#Drop a column
df = df.drop('tip_percent', axis=1)             #axis=1 means drop a column
# df.drop('tip_percent', axis=1, inplace=True)    #Another syntax
```

Set a column as index - See how the display looks like

```python
df = df.set_index('Payment ID')
print(df.head())
df= df.reset_index()
print(df.head())
```

#Conditional Formatting

```python
#Conditional Filtering

print(df['total_bill']<30)		#Boolean Series
print(df[df['total_bill']<30])		#Actual rows where total_bill < 30
print(df[df['gender']=='Male'])
print(df[(df['total_bill']>30) & (df['gender']=='Male')])		# & means AND
print(df[(df['total_bill']>30) &~ (df['gender']=='Male')])		# ~ is Negation (!=)
print(df[(df['total_bill']>30) | (df['tip']>5)])		# | means OR
```

```python

options = ['Sat', 'Sun']
print(df[df['day'].isin(options)])

```

Always Experiment, but Stick to one syntax practically.

One task can be performed in many ways, different syntaxes

- The format or code may change with different syntaxes.

```python
#Gender should be female and tip amount should be $4 or above
syntax1 = df[(df['gender']=='Female')&(df['tip']>=4)]
syntax2 = df.query("gender=='Female' and tip>=4")
syntax3 = df.loc[(df['gender']=='Female')&(df['tip']>=4)]
syntax4 = df.where((df['gender']=='Female')&(df['tip']<=4))
print(syntax1)
print(syntax2)
print(syntax3)
print(syntax4)
```

#Counts, Sorting

```python
gender_counts = df['gender'].value_counts()
print(gender_counts)

sorted_by_bill = df.sort_values(by='total_bill').head(5)
print(sorted_by_bill)

sorted_by_tip_desc = df.sort_values(by='tip',ascending=False).head(5)
print(sorted_by_tip_desc)

sorted_multi = df.sort_values(by=['day','total_bill'],ascending=[True,False]).head(10)
print(sorted_multi)
```
