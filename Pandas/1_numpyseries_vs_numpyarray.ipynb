```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([10,20,30], index=["AliceinWonderland","BobMarley","Cinderella"])
s2 = pd.Series([1,2,3], index=["BobMarley","Cinderella","AliceinWonderland"])
print(s1 + s2)
```

1. Creating a Series from a Python list with Default numeric index

```python
names = ["alice","bob","charlie"]
marks = [34,56,97]
students = pd.Series(data=marks)
print(students)
students = pd.Series(data=marks, index=names)
print(students)
```

2. Creating a Series from a Numpy array

```python
random_ages = np.random.randint(0, 100, 4)
names = ["Alice","Bob","Charlie","Dave"]
people_array = pd.Series(data=random_ages, index=names)
print(people_array)
```

3. Creating a Series from a dictionary

```python
names_ages_dict = {"Alice":21,"Bob":54,"Charlie":87}
print(pd.Series(names_ages_dict))
```

4. Creating a Series for imaginary sales data (Q1 and Q2)

```python
q1 = {'Japan':80,'China':450,'India':200,'USA':250}
q2 = {'Bralzil':80,'China':500,'India':210,'USA':260}

sales_Q1 = pd.Series(q1)
sales_Q2 = pd.Series(q2)

print(sales_Q1)
print(sales_Q1['Japan'])        #Using labelled index
print(sales_Q1[0])              #Using numeric index
```

5. Errors

```python
print(sales_Q1['England'])
print(sales_Q1['india'])
print(sales_Q1[' India'])
```

6. Series operations

```python
print(sales_Q1.keys())
print(sales_Q1 * 2)
print(sales_Q2 / 100)
```

7. Handling mismatched indices (NaN values)

```python
print(sales_Q1 + sales_Q2)
print(sales_Q1.add(sales_Q2, fill_value=0))
```
