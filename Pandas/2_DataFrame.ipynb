```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(101)

scores = np.random.randint(0,101,(4,3))
print(scores)

players = ['Virat','Rohit','MS Dhoni','Hardik Pandya']      #index
matches = ['Match 1','Match 2','Match 3',]                  #column
```

```python
df = pd.DataFrame(data=scores)
print(df)
df = pd.DataFrame(data=scores, index=players)
print(df)
df = pd.DataFrame(data=scores, index=players, columns=matches)
print(df)
```

LOC and ILOC (Indexers, not Functions)

LOC- using numeric index

ILOC- cannot use numeric index, have to use labelled index

Generally we dont see Loc being used, but Iloc we do

```python
print(df)
print(df.info())

print(df.loc['Virat'])              #Row Virat
print(df.loc[['Virat','Rohit']])    #Multiple rows
print(df.loc['Virat','Match 2'])    #Specific cell
print(df.loc[['Virat','MS Dhoni'],['Match 1','Match 3']])   #Subset of Rows and columns
```

```python
print(df.iloc[0])       #First row
print(df.iloc[0:2])     #First two rows
print(df.iloc[0,1])     #First row, second column
print(df.iloc[[0,2],[0,2]])     #Rows 0 & 2, columns 0 & 2
```
