```python
df.head()
df.columns
df.info()
df.describe()
df.describe().transpose()
len(df)
df.isnull().sum()
df.isnull().sum().sum()
df.dtypes
df.dtypes.value_counts()
```

    Data Cleaning

```python
#Handle missing values
df['age'] = df['age'].fillna(df['age'].median()).astype('Int64')
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df.drop(columns='deck', inplace=True, errors='ignore')

#convert categorical columns
df['gender'] = df['gender'].astype('category')
df['embarked'] = df['embarked'].astype('category')

#drop duplicates
df.drop_duplicates(inplace=True)

#Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_')

```

Re-Check

```python
df.info()
df.isnull().sum()
```

2. Find passengers whose age is between 20 and 2.

```python
young_passengers = df[(df['age'] >= 2) & (df['age'] <= 20)]
print(young_passengers)
```

3. Find passengers where gender is female and passenger class is 1.

```python
rich_females = df[(df['gender'] == 'female') & (df['pclass'] == 1)]
print(rich_females)
```

4. Find average age of female and male passengers.


```python
avg_male_age = round(df[df['gender'] == 'male']['age'].mean(),2)
avg_female_age = round(df[df['gender'] == 'female']['age'].mean(),2)

print(f"Average age of males: {avg_male_age}")
print(f"Average age of females: {avg_female_age}")
```

5. Count number of passengers in each passenger class.


```python
no_of_passengers_by_class = df['pclass'].value_counts()
print(no_of_passengers_by_class)
```

6. Separate passengers who survived (survived column contains 0) and who did not survive (survived column contains 1) into two separate dataframes.


```python
survived_passengers = df[df['survived'] == 1]
dead_passengers = df[df['survived'] == 0]
print(f'survived_passengers: {survived_passengers}')
print(f'dead_passengers: {dead_passengers}')
```

7. For the above two dataframes, find number of females and males.


```python
no_of_males_survived = df[(df['gender'] == 'male') & (df['survived'] == 1)]
no_of_females_survived = df[(df['gender'] == 'female') & (df['survived'] == 1)]

no_of_males_dead = df[(df['gender'] == 'male') & (df['survived'] == 0)]
no_of_females_dead = df[(df['gender'] == 'female') & (df['survived'] == 0)]

# Group by gender and survival status, then count
gender_survival_counts = df.groupby(['gender', 'survived'], observed=True).size().unstack(fill_value=0)

# Rename columns for clarity
gender_survival_counts.columns = ['Died', 'Survived']

print(gender_survival_counts)
```

8. Sort the original dataframe on age in descending order and display the top 3 records.


```python
eldest_passengers = df.sort_values(by='age', ascending=False).head(3)
print(f'eldest_passengers: {eldest_passengers}')
```
