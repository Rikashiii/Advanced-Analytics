```python
import numpy as np
from scipy.stats import norm
import math
```

```python
mu0 = 40

sample_means = [80,41,45,-35]

n=100

s=15

alpha = 0.05
```

    #Critical values

norm.cdf() gives Probability, norm.ppf() gives z-score

ppf = percent point function(Inverse of cdf)

Gives z-score corresponding to given probability

e.g Tell me the Z-score such that the area under the curve (left of it) is p

```python

z_crit_two_talied = norm.ppf(1-alpha/2)
z_crit_one_tailed = norm.ppf(1-alpha)
```

```python
print("Critical values")
print("----------------")
print(f"Two-tailed (alpha=0.05: +-{z_crit_two_talied:.3f})")
print(f"One-tailed (alpha=0.05: +-{z_crit_one_tailed:.3f})")
```

```python
#Function to compute Z
def compute_z(xbar):
    return (xbar-mu0)/(s/math.sqrt(n))
```

```python
for xbar in sample_means:
    print(f'=======For sample mean x bar+ {xbar}=======')

    z = compute_z(xbar)
    
```

    #p-values

norm.cdf(z) gives probability to the LEFT of z

So 1-norm.cdf(z) gives probability to the RIGHT of z

For two-tailed: compute to the right and double

e.g Z=2, Right-tail area = 0.0228 = 0.0456

For one-tailed: one sided only

```python


p_two_tailed = 2 * (1 - norm.cdf(abs(z)))
p_one_tailed = 1 - norm.cdf(abs(z))     #imporvement only(right-tail)

print(f"Z-score: {z:.3f}")
```

Two-tailed decision

```python

if abs(z) > z_crit_two_talied:
    two_tail_descision = "Reject H0 (Significant difference)"
else:
    two_tail_descision = "Failed to reject H0 (No significant difference)"

print(f"Two-tailed decision:")
print(f"p-value= {p_two_tailed:.4f}")
print(f"Decision: {two_tail_descision}")
```

One-tailed decision

```python
if z>z_crit_one_tailed:
    one_tail_descision = "Reject H0 (Significant difference)"
else:
    one_tail_descision = "Failed to reject H0 (No significant imporvement)"

print(f"One-tailed test(improvement only):")
print(f"p-value= {p_one_tailed:.4f}")
print(f"Decision: {one_tail_descision}")
```
