    3) Do skewness analysis and plotting for 'on-board Service' column of 'airline passenger service' dataset. Also find CoV for first time versus returning customers for the 'Ease of online booking' column.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
```

```python
obs = df['On-board Service']

mean_obs = obs.mean()
median_obs = obs.median()
skewness_obs = obs.skew()
kurtosis_obs = obs.kurt()

print(f"Mean: {mean_obs}\nMedian: {median_obs}\nSkewness: {skewness_obs}\nKurtosis: {kurtosis_obs}")
```

HistPlot

```python
plt.figure(figsize=(8,6))
sns.histplot(obs, kde=True, bins=30)
plt.axvline(mean_obs, color='r', linestyle='--', label=f'Mean: {mean_obs:.2f}')
plt.axvline(median_obs, color='g', linestyle='-', label=f'Median: {median_obs:.2f}')
plt.title(f'Distribution of On-board Service\nSkewness: {skewness_obs:.2f}, Kurtosis: {kurtosis_obs:.2f}')
plt.xlabel('On-board Service')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

BoxPlot

```python
plt.figure(figsize=(8,6))
sns.boxplot(obs)
plt.tight_layout()
plt.show()
```

Find Coefficient of Variance (CoV) for
    
    First Time vs Returning Customers
for 'Ease of Online Booking'

```python
group1_label = 'First-time'
group2_label = 'Returning'

group1_data = df[df['Customer Type'] == group1_label]['Ease of Online Booking']
group2_data = df[df['Customer Type'] == group2_label]['Ease of Online Booking']
```

```python
#Calculate CoV 
group1_cov = group1_data.std() / group1_data.mean() * 100
group2_cov = group2_data.std() / group2_data.mean() * 100

print(f"CoV for {group1_label}: {group1_cov:.2f}%")
print(f"CoV for {group2_label}: {group2_cov:.2f}%")
```
