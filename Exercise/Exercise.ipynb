```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
# Probability that a passenger dies(Margianl probability)

marginal_prob = df['survived'].value_counts(normalize=True)
marginal_prob
```

```python
# Probability that a passenger was male and survived(joint probability of independent events)

joint_prob = ((df['gender'] == 'male') & (df['survived'] == 1)).mean()

print(joint_prob)
```

```python
#- Probability that a passenger survived given that she waas a female(Conditional/Joint probability of dependent events)
conditional_prob = ((df['gender'] == 'female') & (df['survived'] == 1)).value_counts(normalize=True)
conditional_prob
```

```python
surivied = df.query("gender == 'female' and survived == 1").agg('count')/df.query("gender == 'female'").agg('count')
surivied
```
