# LPP solution using PuLP

```python
from pulp import *
```

Create the problem variable to contain the problem data

```python
model = LpProblem("FactoryProductionProblem", LpMaximize)
```

Create variables for Chairs and Tables

Parameters: Name, Lower limit, Upper limit, Data type

```python
chairs = LpVariable("chairs", 0, None, LpInteger)       #Number of chairs to be produced
tables = LpVariable("tables", 0, None, LpInteger)       #Number of tables to be produced
```

Create maximum objective function (profit)

```python
model += 300 * chairs + 1000 * tables, "Profit"
```

Create constraints

```python
#Wood availability constraint
model += 5* chairs + 10 * tables <= 400, "Wood Constraint"


#Worker time constraint
model += 2* chairs + 6 * tables <= 300, "Time Constraint"

#Adding specific constraint
model += chairs >= 2, "Chairs Constraint"
```

The problem is solved using PuLP's choice of Solver

```python
model.solve()
```

Each of the variables is printed with its resolved optimum value

```python
for v in model.variables():
    print(v.name, "=", v.varValue)

#print objective function value (Total Profit)
print("\nTotal Profit:", value(model.objective))
```
