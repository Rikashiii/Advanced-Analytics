# Student Problem using PuLP

```python
from pulp import *
```

Create the problem variable to contain the problem data

```python
model = LpProblem("StudentProblem", LpMaximize)
```

Create variables for Chairs and Tables

Parameters: Name, Lower limit, Upper limit, Data type

```python
py = LpVariable("Python", 1, 3, LpInteger)               #
sql = LpVariable("SQL", 1, 2, LpInteger)                 #
aa = LpVariable("Advanced Analytics", 1, 5, LpInteger)   #        
H = 120                                                   # Total hours of study availabele                                     
```

Create maximum objective function (profit)

```python
model += py + sql + aa, "Total Credits"
```

Create constraints

```python
#constraint
model += 15*py + 20*sql + 25*aa <= H, "Hours Constraint"
```

The problem is solved using PuLP's choice of Solver

```python
model.solve()
```

Each of the variables is printed with its resolved optimum value

```python
for v in model.variables():
    print(v.name, "=", v.varValue)

#print objective function value (Total Profit)
print("\nTotal Credits:", value(model.objective))
```
