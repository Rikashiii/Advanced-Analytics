```python
import numpy as np
```

```python
sales = [0,5,155,0,518,616]
sales_array = np.array(sales)
```

```python
print(type(sales_array))
print(sales_array.ndim)
print(sales_array.shape)
print(sales_array.size)
print(sales_array.dtype)
```

```python
result_array = np.where(sales_array == 0, "No sales","Some sales")
print(result_array)
```

```python
#show all boolean values
bool_mask = sales_array != 0
print(bool_mask)
```

```python
#filter using the boolean mask
filtered_sales = sales_array[bool_mask]
print(filtered_sales)
```
