   We cannot append an array, once its made.
   To do that we have to create a new array.

```python
import numpy as np
arr = np.arange(0,10)
new_arr_single = np.append(arr, 10)
new_arr_multiple = np.append(arr, [10,11,12])
print(arr)
print(new_arr_single)
print(new_arr_multiple)
```

#Shallow copy: Suppose we have array a, then we do b =a[0:5]

-> b will not have serparate memory, instead b will point to the array a

b->a->memory

#Deep copy: Suppose we have array a, then we do b=a.copy()
-> b will have separate memory, not a's memory

b->memory   a->memory

if we change b, it will not impact a ... and vice-versa
