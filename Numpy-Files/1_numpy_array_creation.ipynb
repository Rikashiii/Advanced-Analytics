These are special kinds of arrays like full or zeros, ones, eyes

```python
#special arrays 
zeros = np.zeros((3, 4))  # 3x4 array of zeros
ones = np.ones((2, 5))    # 2x5 array of ones
identity = np.eye(4)      # 4x4 identity matrix
print(zeros)
print(ones)

empty = np.empty((2, 3))  # 2x3 empty array
full = np.full((3, 3), 7)  # 3x3 array filled with 7
print(empty)
print(full)
print(identity)
```

Arange vs Linspace function
- Arange exludes upper end while linspace includes.
- (_,_,x) x is step in arange (How many elements we want to skip)
- (_,_,x) produces x evenly spaced values in Linspace(How many elements we want)

```python
a = np.arange(1,10,3)
l = np.linspace(1,10,3)
lb = np.linspace(1,10,6, dtype=int)
print(a, l, lb)
```
