#creating multi-dimensional arrays as a "structure"

```python
import numpy as np
student = np.array([
    ('Alice',95,20),
    ('Bob',88,21),
    ('Charlie',92,19)
], dtype= [('name','U10'),('marks','i4'),('age','i4')
])
```

```python
print(student)
print(student['name'])
print(student['marks'])
print(student['marks'].mean())
print(student[student['name']=='Alice']['marks'])       #dtype32
print(student[student['name']=='Alice']['marks'][0])
```
