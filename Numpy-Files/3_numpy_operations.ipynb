Numpy Operations

```python
import numpy as np
```

```python
arr = np.arange(0,10)
```

```python
#Basic arithmatics
print(arr + arr)
print(arr - arr)
print(arr * arr)
print(arr / arr)
print(arr ** 3)
print(np.sqrt(arr))
```

```python
#Summary statistics
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.var())    #variance
print(arr.std())    #standard deviation
```

```python
#Reversing an array
arr = np.array([1,2,3,4,5])
reversed_arr = arr[::-1]
print(reversed_arr)
```

```python
#Sorting an array
arr = np.array([5,2,8,1,3])
sorted_arr = np.sort(arr)
print(sorted_arr)

#For descending
#there is no option of descending in numpy sir!!! please use pandas!!
sorted_arr = np.sort(arr)[::-1]     #Reverse the party
print(sorted_arr)
```
