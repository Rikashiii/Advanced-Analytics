Paired t-test: A company has launched some offer
#H0: Pre and post usage is the same (i.e offer is not effective)

```python
from scipy.stats import ttest_rel
from math import sqrt

import pandas as pd
```

```python
df = pd.read_csv("cust_seg.csv")
usage_data = df[["pre_usage","Post_usage_1month"]]
usage_data = usage_data.sample(n=30)
```

```python
tstat, pval = ttest_rel(usage_data["pre_usage"],usage_data["Post_usage_1month"])

alpha = 0.05

if pval < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
