    One sample T-test

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('Mall_customers.csv')
```

```python
age_data = data['Age']

age_data = age_data.sample(n=30, replace=False)
pop_avg_age = 40 # population average age       #H0: Mall customers have an average age of 40

t_statistics, p_value = stats.ttest_1samp(age_data, pop_avg_age)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```
