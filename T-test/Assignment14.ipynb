For combined Marathon datasets, find if younger (age <= 30) and older (age > 30)
	athletes have a similar average pace.

(H0: There is no statistically significant diff between the pace of the younger and older atheletes)


```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df15 = pd.read_csv("marathon_results_2015.csv")
df16 = pd.read_csv("marathon_results_2016.csv")
df17 = pd.read_csv("marathon_results_2017.csv")

df = pd.concat([df15, df16, df17])
df.head()
```

```python
# 1. CONVERT TIME STRINGS TO NUMBERS
# 'coerce' turns messy data into NaN (Not a Number) so it doesn't crash
df['Pace_Numeric'] = pd.to_timedelta(df['Pace'], errors='coerce').dt.total_seconds()

# 2. FILTER & SPLIT (Using the NEW numeric column)
# We drop NaNs here to ensure clean input for the stats function
younger = df[df['Age'] <= 30]['Pace_Numeric'].dropna()
older = df[df['Age'] > 30]['Pace_Numeric'].dropna()

# 3. RUN THE TEST
t_stat, p_val = stats.ttest_ind(younger, older)

print(f"T-stat: {t_stat}")
print(f"P-value: {p_val}")
```

Perform the t-test
