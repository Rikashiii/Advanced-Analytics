    Matches against different oppositions

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('MS_Dhoni_ODI_record.csv')
```

```python
opposition_counts = df['opposition'].value_counts()
print(opposition_counts)
```

Matplotlib bar plot

```python
plt.Figure(figsize=(8,5))
plt.bar(opposition_counts.index, opposition_counts.values, color='green')
plt.title('No of matches against different oppositions (Matplotlib)')
plt.xticks(rotation=90)
plt.show()
```

Seaborn Version

```python
sns.countplot(x='opposition', data=df)
plt.title('No of matches against different oppositions (Seaborn)')
plt.xticks(rotation=90)
plt.show()
```
