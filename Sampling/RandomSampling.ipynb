```python
import pandas as pd
import numpy as np
```

```python
df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")
df = pd.concat([df_2015, df_2016, df_2017])
original_row_count = len(df)
```

::7  means from the start to the end, pick every 7th record,

: means all columns

```python
sample_df = df.iloc[::7,:]
sample_row_count = len(sample_df)

print(df.head())
print(sample_df.head())
print(original_row_count)
print(sample_row_count)
```

```python
#Stratified random sampling
total_males = len(df[df['M/F'] == 'M'])
total_females = len(df[df['M/F'] == 'F'])

sample_size = int(0.1*original_row_count)
sample_males = int(0.1*total_males)
sample_females = int(0.1*total_females)
print(f'sample size: {sample_size},\n sample males: {sample_males},\n sample females: {sample_females}')
```

```python
#Generate proportionate samples of male and female athelets
male_sample = df[df['M/F'] == 'M'].sample(n=sample_males,random_state=42)
female_sample = df[df['M/F'] == 'F'].sample(n=sample_females,random_state=42)
print(len(male_sample),len(female_sample))
```

```python
#combine the male and female samples into a single sample dataframe
sample_df = pd.concat([male_sample,female_sample])
sample_row_count = len(sample_df)
print(df.head())
print(sample_df)


print(sample_row_count)
```

```python
print(f"No of male athelets in sample: {len(sample_df[sample_df['M/F'] == 'M'])}")
print(f"No of female athelets in sample: {len(sample_df[sample_df['M/F'] == 'F'])}")
```

    Cluster Sampling

Find th unique countries the atheletes are from

```python
unique_countries = df['Country'].unique()
num_couintries = len(unique_countries)
print(unique_countries,num_couintries)
```

```python
#Randomly select half of the countries 
np.random.seed(42)
```

#choice() function in numpy generates a randmom sample from an array

#replace=False means do not allow the same country twice or more

```python
selected_countries = np.random.choice(unique_countries, size=num_couintries//2, replace=False)
print(selected_countries)
```

```python
#create a new dataframe containing all records of the selected countries
sample_df = df[df['Country'].isin(selected_countries)]
sample_row_count = len(sample_df)
print(df.head(),sample_df.head())
```
