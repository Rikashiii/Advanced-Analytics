```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('tips.csv')
```

Scatter Plot of 'total_bill' vs 'tip' 

    Tells us how these are realted to each other

```python
plt.Figure(figsize=(6,4))
#divide 'total_bill' into 10 different categories, try changing bins
plt.scatter(df['total_bill'],df['tip'],color='orange',edgecolor='black')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

Same with seaborn, but also with Regression Line

```python
plt.figure(figsize=(6,4))
sns.regplot(x='total_bill',y='tip',data=df, color='Red')
plt.title('Total Bill vs Tip with Regression Line')
plt.show()
```
