```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('tips.csv')
```

1. Histogram of total_bill - Shows the distribution of a single variable (here, total bill)

plt.Figure(figsize=(_,_))       #sets size of the canvas

```python
plt.Figure(figsize=(6,4))
#divide 'total_bill' into 10 different categories, try changing bins
plt.hist(df['total_bill'],bins=51,color='purple',edgecolor='lime')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

Same plot, using Seaborn

Estimation- shows a continuous line over the histogrm


```python
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'],bins=52, kde=True, color='Red')      #kde = Kernel Density

plt.title('Distribution of Total Bill')
plt.show()
```
