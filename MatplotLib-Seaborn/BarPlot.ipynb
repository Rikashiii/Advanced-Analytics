```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('tips.csv')
```

Bar plot of average tip by gender

Grouping automatically creates an index on the grouped column, in this case, Gender.

```python
avg_tip_gender = df.groupby('gender')['tip'].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_tip_gender.index,avg_tip_gender.values, color=['Maroon','Pink'])        #index= Gender
plt.title('Avg Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```

Same shit using Seaborn

```python
avg_tip_gender = df.groupby('gender')['tip'].mean().reset_index()
plt.figure(figsize=(6,4))
sns.barplot(data=avg_tip_gender, x='gender',y='tip', palette=['Maroon','Pink'])
plt.title('Avg Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```
