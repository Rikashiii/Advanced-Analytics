```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#1

```python
# Do outlier analysis on total bill column of tips dataset.
data_array = pd.read_csv("tips.csv")['total_bill'].values

q1 = np.percentile(data_array, 25)  # 25th percentile = Q1
q3 = np.percentile(data_array, 75)  # 75th percentile = Q3
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = data_array[(data_array < lower_fence) | (data_array > upper_fence)]

print(f"Q1: {q1}")
print(f"Q2: {np.median(data_array)}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Lower Fence: {lower_fence}")
print(f"Upper Fence: {upper_fence}")
print(f"Outliers: {outliers}")
```

```python
#create box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=data_array,orient='h')
plt.title('Box Plot of Total Bill Amounts')
plt.xlabel('Total Bill ($)')
plt.show()
```
