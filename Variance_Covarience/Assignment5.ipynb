```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('social_media_usage.csv')
```

```python
#Do correlation and covanance analysis and visualization on the 'social media' usage separately for facebook and instagram users.
# Columns to be used: daily minutes per day and posts per day.

df_facebook = df[df['App'] == 'Facebook']
df_instagram = df[df['App'] == 'Instagram']
```

Calculate correlation

```python
corr_facebook = df[['Daily_Minutes_Spent', 'Posts_Per_Day']].corr()
corr_instagram = df[['Daily_Minutes_Spent', 'Posts_Per_Day']].corr()
```

Calculate Covariance

```python
cov_facebook = df[['Daily_Minutes_Spent', 'Posts_Per_Day']].cov()
cov_instagram = df[['Daily_Minutes_Spent', 'Posts_Per_Day']].cov()
```

```python
print(f'correlation Facebook:\n {corr_facebook}')
print(f'covariance Facebook:\n {cov_facebook}')
print(f'correlation Instagram:\n {corr_instagram}')
print(f'covariance Instagram:\n {cov_instagram}')
```

    Visualization (Heatmap)

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Facebook Correlation Heatmap
sns.heatmap(corr_facebook, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title("Facebook Correlation")

# Instagram Correlation Heatmap
sns.heatmap(corr_instagram, annot=True, cmap='YlGnBu', ax=axes[1])
axes[1].set_title("Instagram Correlation")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Facebook Correlation Heatmap
sns.heatmap(cov_facebook, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title("Facebook Covariance")

# Instagram Correlation Heatmap
sns.heatmap(cov_instagram, annot=True, cmap='YlGnBu', ax=axes[1])
axes[1].set_title("Instagram Covariance")

plt.tight_layout()
plt.show()
```
