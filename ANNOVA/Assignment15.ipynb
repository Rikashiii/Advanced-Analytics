Fir StudentsPerformance.csv, using ANOVA find if there is a significant difference in Maths marks depending on the level of parental education.

For ANNOVA, there is no need to sample.

Use the full dataset

```python
from scipystats import f_oneway
import pandas as pd
```

```python
df = pd.read_csv("StudentsPerformancecsv")
```

```python
print(df['math score'].value_counts())
```

ANOVA requires separating the numeric data into groups based on the category.

```python
# This creates a list of arrays, where each array contains math scores for a specific education level
groups = [group['math score'] for level, group in df.groupby('parental level of education')]

f_stat, p_val = f_oneway(*groups)
```

 INTERPRET RESULTS


```python
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value:     {p_val:.4e}")
```

```python
# Hypothesis Testing 
alpha = 0.05

if p_val < alpha:
    print("Reject the Null Hypothesis.")
    print("Conclusion: There IS a significant difference in math scores depending on parental education.")
else:
    print("Fail to reject the Null Hypothesis.")
    print("Conclusion: There is NO significant difference in math scores.")
```
