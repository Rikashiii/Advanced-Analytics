```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
```

Combine datasets

```python
result20 = pd.read_csv('Men ODI Team Match Results - 20th Century.csv')
result21 = pd.read_csv('Men ODI Team Match Results - 21st Century.csv')
result_df = pd.concat([result20, result21])
result_df.head()
```

Historical winrate of India

```python
india_matches_df = result_df[result_df['Match'].str.contains('India', case=False, na=False)]    #case sens OFF, na val OFF


india_wins = india_matches_df[india_matches_df['Result'] == 'Won'].shape[0]
print(india_wins)
total_matches = india_matches_df.shape[0]
print(total_matches)

india_winrate = (india_wins / total_matches) * 100
print(f"India winrate: {india_winrate:.2f}%")
```

Now India is scheduled to play 60 ODI matches over the next 3 years.

(a) What is the probability that India wins exactly 35 out of these 60 matches?

(b) What is the probability that India wins more than 40 matches?

(c) What is the probability that India wins at least 45 matches?

(d) What is the most likely number of wins India will have in 60 matches?

```python
#(a) What is the probability that India wins exactly 35 out of these 60 matches?
a = binom.pmf(35, 60, india_winrate/100)
print(a)
b = 1 - binom.cdf(40, 60, india_winrate/100)
print(b)
c = 1 - binom.cdf(44, 60, india_winrate/100)
print(c)
#(d) What is the most likely number of wins India will have in 60 matches?
d = int((60+1)*india_winrate/100)
print(d)
```
