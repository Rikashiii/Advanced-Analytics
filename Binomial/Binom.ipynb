# If in general, 66% students like python, find the probability that exactly 3 out of 5 students will like Python - Binomal Distribution problem

```python
from scipy.stats import binom
```

```python
# Given values
n = 5 # Total number of students
k = 3 # Number of students who like Python
p = 0.66 # Probability that a student likes Python
```

```python
# Calculate binomial probability
probability = binom.pmf(k,n,p)
print(probability)
```
