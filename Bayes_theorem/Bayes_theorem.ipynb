```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
total_passengers = len(df)
```

```python
survived_count = len(df[df['survived'] == 1])
```

```python
female_count = len(df[df['gender']==1])
```

```python
female_survived_count = len(df[(df['survived']==1) & (df['gender']==1)])
```

```python
P_survived = survived_count / total_passengers
```

```python
P_female = female_count / total_passengers
P_female
```

```python
P_female_survived = female_survived_count / survived_count
P_female_survived
```
