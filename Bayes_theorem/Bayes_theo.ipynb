```python
import pandas as pd
```

```python
df = pd.read_csv('loan_approval_dataset.csv')
```

```python
N = len(df)
```

```python
A = len(df[df['loan_status'] == 'Approved'])
```

```python
B = len(df[df['cibil_score']>=700])
```

```python
# A and B = Approved AND GOOD Credit
A_and_B = len(df[(df['loan_status'] == 'Approved') & (df['cibil_score']>=700)])
```

```python
P_A = A/N
P_B = B/N
P_B_given_A = A_and_B/A
```

```python
# Bayes Theorem
P_A_given_B = P_B_given_A * P_B / P_A
```

```python
print("Total_Records: ", N )
print("Approved: ", A)
print("Good Credit: ", B)
print("Approved AND Good Credit: ", A_and_B)
print("Probability of Approved: ", P_A)
print("Probability of Good Credit: ", P_B)
print("Probability of Approved given Good Credit: ", P_B_given_A)
print("Probability of Good Credit given Approved: ", P_A_given_B)
```

```python
# Bad Credit count
bad_credit = N - B
```

```python
# Approved AND Bad Credit
A_and_notB = A - A_and_B
```

```python
# P(Loan Approved | Bad Credit)
P_A_given_notB = (A_and_notB/ A) * (P_A /(bad_credit/N))
```

```python
print("P(Loan Approved | CIBIL Score = 800): ", P_A_given_B) #Good Credit
print("P(Loan Approved | CIBIL Score = 500): ", P_A_given_notB) #Bad Credit
```
